Scaling of Variables 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
data = pd.read_csv('https://raw.githubusercontent.com/arijitbose/Feature_engineering/main/Travel.csv')
data
```

```python
data.drop('CustomerID', axis=1, inplace=True)
```

```python
category = [feature for feature in data.columns if data[feature].dtype!='O']
category
```

```python
data[category]
```

```python
for feature in data[category]:
    mean = data[feature].mean()
    data[feature] = data[feature].fillna(mean)
```

```python
data.isnull().sum()
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import RobustScaler
from sklearn.preprocessing import MinMaxScaler
```

```python
std = StandardScaler()
```

```python
pd.DataFrame(std.fit_transform(data[category]), columns=['ProdTaken',
 'Age',
 'CityTier',
 'DurationOfPitch',
 'NumberOfPersonVisiting',
 'NumberOfFollowups',
 'PreferredPropertyStar',
 'NumberOfTrips',
 'Passport',
 'PitchSatisfactionScore',
 'OwnCar',
 'NumberOfChildrenVisiting',
 'MonthlyIncome'])
```

```python
robust = RobustScaler()
```

```python
data = pd.read_csv('https://raw.githubusercontent.com/arijitbose/Feature_engineering/main/Travel.csv')
data
```

```python
categorical_variables = [feature for feature in data.columns if data[feature].dtype!='O']
categorical_variables
```

```python
data[categorical_variables]
```

```python
dt = data[categorical_variables].copy()
```

```python
pd.DataFrame(robust.fit_transform(dt), columns=['CustomerID',
 'ProdTaken',
 'Age',
 'CityTier',
 'DurationOfPitch',
 'NumberOfPersonVisiting',
 'NumberOfFollowups',
 'PreferredPropertyStar',
 'NumberOfTrips',
 'Passport',
 'PitchSatisfactionScore',
 'OwnCar',
 'NumberOfChildrenVisiting',
 'MonthlyIncome'])
```

```python
dd = data[categorical_variables].copy()
```

```python
min_max = MinMaxScaler()
min_max
```

```python
pd.DataFrame(min_max.fit_transform(dd), columns=['CustomerID',
 'ProdTaken',
 'Age',
 'CityTier',
 'DurationOfPitch',
 'NumberOfPersonVisiting',
 'NumberOfFollowups',
 'PreferredPropertyStar',
 'NumberOfTrips',
 'Passport',
 'PitchSatisfactionScore',
 'OwnCar',
 'NumberOfChildrenVisiting',
 'MonthlyIncome'])
```

```python
from sklearn.preprocessing import Normalizer
```

```python
norm  = Normalizer(norm='l1')
norm
```

```python
dp = data[categorical_variables].copy()
dp
```

```python
for feature in dp.columns:
    mean = dp[feature].mean()
    dp[feature] = dp[feature].fillna(mean)
```

```python
dp.isnull().sum()
```

```python
pd.DataFrame(norm.fit_transform(dp), columns=['CustomerID', 'ProdTaken', 'Age', 'CityTier', 'DurationOfPitch',
       'NumberOfPersonVisiting', 'NumberOfFollowups', 'PreferredPropertyStar',
       'NumberOfTrips', 'Passport', 'PitchSatisfactionScore', 'OwnCar',
       'NumberOfChildrenVisiting', 'MonthlyIncome'])
```

```python
from sklearn.preprocessing import MaxAbsScaler
```

```python
max_ab = MaxAbsScaler()
max_ab
```

```python
ds = data[categorical_variables].copy()
ds
```

```python
for feature in ds.columns:
    mean = ds[feature].mean()
    ds[feature] = ds[feature].fillna(mean)
```

```python
ds.isnull().sum()
```

```python
pd.DataFrame(max_ab.fit_transform(ds), columns=['CustomerID',
 'ProdTaken',
 'Age',
 'CityTier',
 'DurationOfPitch',
 'NumberOfPersonVisiting',
 'NumberOfFollowups',
 'PreferredPropertyStar',
 'NumberOfTrips',
 'Passport',
 'PitchSatisfactionScore',
 'OwnCar',
 'NumberOfChildrenVisiting',
 'MonthlyIncome'])
```
